#### Extracting features and Target variables

<b>Target variable</b>:
True if the user is active for more than 6 months (6x30 days = 180 days).<br>

Step 1: For a user since his/her last_trip_date the next 29 days would be considered as active user. Creating column last_active_date.<br>
Step 2: If the difference between signup_date and last_active_date is greater than 180 days, consider user retained. Creating column six_month_retaintion.


```python
from datetime import timedelta
dfFeatured = dfFilled.copy()
dfFeatured['last_active_date'] = dfFeatured.last_trip_date + timedelta(days=29)
dfFilled.head()
```

```python
dfFeatured['active_days']=(dfFeatured.last_active_date-dfFilled.signup_date).dt.days
dfFeatured['six_month_retaintion']=np.int64(dfFeatured.active_days>=180)
dfFeatured.head(10)
```

### Algorithm selection
<b>Working with Nulls</b>: The avg_rating_of_driver have sizable chuck of nulls but none of the sklearn algorithm accepts null values. So the option here is to 
- Either drop the column, leading to loss in information.
- Fill the nulls with some strategy.<br>

<p>Here we used sklearn's experimental feature that takes into consideration rest of the features to gauge the missing value in one feature. This is will only be effective when filling one column while other columns are non-nulls.</p>

<b>Moldel selection</b>: We try two different algorithms Random Forest (RF) and Support Vector Machines (SVM). RF are general purpose and are mostly perform well. We would want to compare SVM since it is usually yields good reusults for simple classification problems.

```python
import sklearn.metrics as metrics 
from sklearn.model_selection import GridSearchCV, train_test_split

def plot_roc_auc(X_test,y_test, cv):
    print('Test Score: ',cv.score(X_test,y_test))

    probs = cv.predict_proba(X_test)
    preds = probs[:,1]
    fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
    roc_auc = metrics.auc(fpr, tpr)

    plt.title('ROC')
    plt.plot(fpr, tpr, 'b', label='AUC = %0.2f' % roc_auc)
    plt.legend(loc='lower right')
    plt.plot([0,1],[0,1],'r--')
    plt.xlim([0,1])
    plt.ylim([0,1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    
def trainModel(pipeline, dfEncoded, parameters, X_train, X_test, y_train, y_test):
    cv = GridSearchCV(pipeline, cv=5, param_grid=parameters, iid=True)
    cv.fit(X_train, y_train)
    print('Best Estimator:', cv.best_estimator_)
    plot_roc_auc(X_test,y_test, cv)
    return cv
```
