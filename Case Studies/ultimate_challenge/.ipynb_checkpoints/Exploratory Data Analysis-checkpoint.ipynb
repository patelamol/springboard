```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
logins_df = pd.read_json('logins.json', dtype={'login_time':'pd.datetime'})
print('Shape before setting index. ',logins_df.shape)
logins_df = logins_df.set_index('login_time')
print('Shape after setting index. ',logins_df.shape)
logins_df['isLogin'] = True
logins_df.head()
```

```python
logins15min = logins_df.resample('15T').count()
logins15min['Date'] = logins15min.index.date
logins15min['Hour'] = logins15min.index.hour
logins15min['Time'] = logins15min.index.time
logins15min['Day'] = logins15min.index.weekday
print(logins15min.head())
print(logins15min.info())
```

The login time shows a saw tooth pattern, with regular peaks. To know how often these peaks occur we need to zoom into a small section.<br>
Also, note there seem to be couple of anomalies in March and April, where you can see large spikes.

```python
logins15min.reset_index().plot(x='login_time',y='isLogin',figsize=(20,10))
plt.show()
```

Looking at logins in a 24 hours cycle. Looks like the logins drops around morning commute times but spikes up around lunch times. Suprisingly the logins are constantly higher in the night vs day time.

```python
loginGrped = logins15min.groupby('Time')['isLogin'].sum()
loginGrped.plot(figsize=(20,10))
plt.show()
```

Looking into weekly lens, its clear most of the logins happen over the weekend. With a ramp up from Monday to Saturday, with some drop on Sunday.

```python
loginGrped = logins15min.groupby('Day')['isLogin'].sum()
loginGrped.plot(figsize=(20,10))
plt.show()
```
